```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from pathlib import Path


def find_data_file():
    candidates = [
        Path.cwd() / "data" / "UAVIDS-2025.csv",
        Path.cwd().parent / "data" / "UAVIDS-2025.csv",
        Path("data/UAVIDS-2025.csv"),
    ]
    for candidate in candidates:
        if candidate.exists():
            return candidate
    raise FileNotFoundError("Could not find data/UAVIDS-2025.csv. Check that the file exists in the project data folder.")


data_path = find_data_file()
df = pd.read_csv(data_path)

print("Loaded dataset from:", data_path)
print(df.head())
print("\nShape:", df.shape)
print("\nColumns:", list(df.columns))
```
